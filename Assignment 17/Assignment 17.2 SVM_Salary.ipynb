```python
# Encoding of both train data and test data
```

```python
cat_col = df_Train.select_dtypes(object).columns
cat_col
```

```python
oe =OrdinalEncoder()
df_Train[cat_col] = oe.fit_transform(df_Train[cat_col])
```

```python
df_Train
```

```python
cat_test = df_Test.select_dtypes(object).columns
cat_test
```

```python
df_Test[cat_test] = oe.fit_transform(df_Test[cat_test])
```

```python
df_Test
```

```python
# Training and Testing Data
```

```python
xtrain = df_Train.iloc[:,:-1]
ytrain = df_Train['Salary']
```

```python
xtest = df_Test.iloc[:,:-1]
ytest = df_Test['Salary']
```

```python
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtrain
```

```python
xtest = sc.fit_transform(xtest)
xtest
```

```python
# Model Buliding
```

```python
svc = SVC()
```

```python
svc.fit(xtrain,ytrain)
ypred = svc.predict(xtest)
```

```python
print(classification_report(ytest,ypred))
```
