```python
# Hirerichal Clustering
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc = StandardScaler()
y = sc.fit_transform(x)
y
```

```python
import scipy.cluster.hierarchy as sch
```

```python
lk = sch.linkage(y,method='complete')
dendro = sch.dendrogram(lk)
plt.show()
```

```python
#KMEANS
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
```

```python
ab = km.fit_predict(x)
```

```python
km.cluster_centers_
```

```python
x
```

```python
plt.scatter(y[:,0],y[:,2],c=ab,cmap='viridis')
plt.show()
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=3)
model.fit(x)
model.labels_
```

```python
x['Cluster']= model.labels_
x
```

```python
x.groupby(x.Cluster).mean()
```
