```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = (200)
plt.rcParams['figure.figsize'] =(15,20)
from sklearn.preprocessing import OrdinalEncoder ,LabelEncoder
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\Abhishek\\Downloads\\fraud_check.csv',encoding='latin1')
df
```

```python
df.describe()
```

```python
# Encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
df_cat_col = df.select_dtypes([object]).columns
df_cat_col
```

```python
df[df_cat_col] = oe.fit_transform(df[df_cat_col])
df
```

```python
df['Tax'] = df['Taxable.Income'].map(lambda x : 'Risky' if x <=30000
                                    else 'Good')
```

```python
df.head(5)
```

```python
df.drop(columns='Taxable.Income',inplace=True)
```

```python
# Model Building
```

```python
x = df.iloc[:,:-1]
y = df['Tax']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
print(f"Training Accuracy:{dt.score(xtrain,ytrain)}\n Testing Accuracy:{dt.score(xtest,ytest)}")
```

```python
# Decision Tree Model
```

```python
tree.plot_tree(dt, feature_names=x.columns, filled=True)
plt.show()
```

```python
# HyperparameterTuning
```

```python
params = {
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':[2,3,4,5],
    
}
```

```python
grid = GridSearchCV(DecisionTreeClassifier(),param_grid=params,verbose=1)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
dt = DecisionTreeClassifier(criterion='gini',max_depth=4,splitter='random')
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
print(f"Training Accuracy:{dt.score(xtrain,ytrain)}\n Testing Accuracy:{dt.score(xtest,ytest)}")
```

```python
tree.plot_tree(dt)
plt.show()
```
