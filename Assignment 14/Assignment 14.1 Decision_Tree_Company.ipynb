```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = (200)
plt.rcParams['figure.figsize'] =(15,20)
from sklearn.preprocessing import OrdinalEncoder ,LabelEncoder
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\Abhishek\\Downloads\\company_Data.csv',encoding='latin1')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Sales'].mean()
```

```python
df['High']=df.Sales.map(lambda x: 1 if x > 7.49
                       else 0)
```

```python
le  = LabelEncoder()
```

```python
df['ShelveLoc'] = le.fit_transform(df['ShelveLoc'])
df['Urban'] = le.fit_transform(df['Urban'])
df['US'] = le.fit_transform(df['US'])
df
```

```python
plt.subplot(2,2,1)
sns.countplot(df['Urban'])

plt.subplot(2,2,2)
sns.countplot(df['US'])

plt.subplot(2,2,3)
sns.countplot(df['High'])
```

```python
x = df.iloc[:,1:11]
y = df['High']

x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

```python
tree.plot_tree(dt, feature_names=x.columns, 
               class_names=['Sale is High','Sale is Low'],
              filled=True)
plt.show()
```

# Hyperparametr Tuning

```python
dt= DecisionTreeClassifier()

param= {
    'criterion': ['gini','entropy'],
    'splitter':['random','best'],
    'max_depth':[1,2,3,4],
     'min_samples_split':[2,3,4,6,8]
}
```

```python
grid = GridSearchCV(DecisionTreeClassifier(),param_grid=param,verbose=1)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
dt = DecisionTreeClassifier(criterion='gini',min_samples_split=3,
                           max_depth=4, splitter='random')

dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

```python
print(classification_report(ytest,ypred))
```

```python
tree.plot_tree(dt)
plt.show()
```
