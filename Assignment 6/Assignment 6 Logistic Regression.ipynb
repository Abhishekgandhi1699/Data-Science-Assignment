# train_test_split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,LabelEncoder
```

```python
x = df.iloc[:,0:-1]
y = df['y']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
```

```python
lg.fit(xtrain,ytrain)
ypred = lg.predict(xtest)
```

```python
len(xtest)
```

```python
len(ypred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print('act val :',ytest.values[:25])
print('pred val:',ypred[:25])
```

```python
print(classification_report(ytest,ypred))
```

```python
confusion_matrix(ytest,ypred)
```

# Threshold Selection

```python
proba=lg.predict_log_proba(xtest)[:,1]
```

```python
proba
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score
```

```python
auc = roc_auc_score(ytest,proba)

fpr,tpr, threshold = roc_curve(ytest,proba)

plt.plot(fpr,tpr,label = '(AUC = %0.2f)'%auc)
plt.legend()
```

```python
auc
```

```python
threshold
```

```python
accuracy = []

for i in threshold:
    ypred = np.where(proba>i,1,0)
    accuracy.append(accuracy_score(ytest,ypred))

final_threshold = pd.concat([pd.Series(threshold),pd.Series(accuracy)], axis=1)
final_threshold.columns = ['Threshold','Accuracy']
```

```python

final_threshold.sort_values(by='Accuracy', ascending=False)
```
