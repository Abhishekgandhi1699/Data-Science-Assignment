```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

# Question 1

```python
df=pd.read_csv('C:\\Users\\Abhishek\\Downloads\\book.csv',encoding='latin1')
df
```

```python
df.info()
```

```python
df.describe()
```

# Confidence =70% and Support =10%

```python
scores=apriori(df,min_support=0.1,use_colnames=True)
scores
```

```python
model=association_rules(scores,metric='confidence',min_threshold=0.7)
model
```

```python
plt.scatter(model['support'],model['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

# Confidence=80% and Support =5%

```python
scores1=apriori(df,min_support=0.1,use_colnames=True)
scores1
```
